# Local Question Answer Chatbot using RAG 

## Step1 : Loading Libraries 

```python
import os 

#load pdf document 
from langchain_community.document_loaders import PyPDFLoader

# load CSV document 
from langchain_community.document_loaders import CSVLoader

# load text document 
from langchain_community.document_loaders import TextLoader

#vector databse used to store embedding 
from langchain_community.vectorstores import Chroma

# converting text to vector 
from langchain_community.embeddings import HuggingFaceEmbeddings

#ollama 
from langchain_community.llms import ollama 

## import ollama

import warnings
warnings.filterwarnings("ignore")
```

## Step 2 : Folder Path 

```python
DATA_FOLDER = "data"
VECTOR_DB_FOLDER = "vector_db"
```

## Step 3: load the document from file folder 


```python
def load_document(): 
    documents = []

    print("\n Scanning from file.....")

    for file in os.listdir(DATA_FOLDER): 
        file_path = os.path.join(DATA_FOLDER,file)

        ##load pdf file 
        if file.endswith(".pdf"): 
            print("Loading pdf file ....")

            loader = PyPDFLoader(file_path)
            docs = loader.load()


            documents.extend(docs)

            print("PDF loaded sucessfully !!! with. Pages: ", len(docs))

    return documents

docs = load_document()

            
```

## Step 3 Split the documents into chunks 

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

def split_documents(documents):
    print("Spliting documents into chunks:")
    
    text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 100)

    chunks = text_splitter.split_documents(documents)

    print("Documents splitted into chunks sucessfully!!! ")
    print("Total chunks created  : ", len(chunks))

    ## show the preview 
    print("\n Sample chunk preview:")
    print(chunks[0].page_content[:100])

    return chunks 

chunks = split_documents(docs)   
    
```

### Step 4: Create a vector db using embeddings 


```python
def create_vector_database(chunks):

    print("Loading embedding models")
    embeddings = HuggingFaceEmbeddings()
    print("Embedding model of huggingface loaded sucessfully \n")
    

    ## create a vector database 
    print("creating chroma database: ")
    db = Chroma.from_documents(chunks, embeddings,persist_directory=VECTOR_DB_FOLDER )

    ##store data locally 
    print("Saving data to disk:")
    db.persist()
    print("Data saved sucessfully !!!. Total chunks ", len(chunks))

    ## show the preview 
    print("\n Sample chunk preview:")
    print(chunks[1].page_content[:300])
    

    return db 

db = create_vector_database(chunks)

```

### Step 5: Loading the vector database 

```python
def load_vector_database():
    embeddings = HuggingFaceEmbeddings()
    
    db = Chroma(
        persist_directory=VECTOR_DB_FOLDER,
        embedding_function = embeddings
    )

    return db 

load_vector_database()
    
```

### Step 6: Chatbot interaction loop 



```python
from langchain_ollama import OllamaLLM

def start_chatbot():

    ## load vector database 

    db = load_vector_database()
    llm = OllamaLLM(model="llama3")


    while True: 

        #question 
        query = input("Ask any question :")

        if query.lower() == "exit": 
            print("Chatbot stopped")
            break

        ## retrieve relevent documents 
        docs = db.similarity_search(query, k =3)

        context = ""

        for i, doc in enumerate(docs): 
            context += doc.page_content + "\n"

        prompt = f"""

        You are a document question answering assitant 

        Rules: 
        1. Answer only from the provided context 
        2. Dot not say cannot access file 
        3. Donot copy the context word by word unless needed  
        4. Give short and clear answer 
        5. If the answer is not in context then reply exactly answer not found in document 

        Context : 
        {context}

        Question: 
        {query}

        Final Answer: 
        
        """

        response = llm.invoke(prompt)

        print(response)


start_chatbot()


        

        
            
```
