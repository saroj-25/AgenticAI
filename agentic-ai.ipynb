## Agentic AI 

- It is AI system where we not only generate text but also considers goals, reasons about next step, check results, works until you outcome is achieved. 
- It uses LLM as brain, retrival memory, and tools used to perform specific tasks. 


## How it works 

- Input: 
  User gives their goal 

- Reason: 
    The model interprates the goal 

- Act: 
    Model calls for tool or performs steps of operation 

- Observe: 
    The system reads the result 

- Repeat: 
    It keep on going until all the task are completed. 

This loop is what makes system agentic 


## Agentic AI vs LLM(Large Language Models)

- A plain LLM maily pedicts next token in text and return the result based on prompts and consider the context. 
- An AI agent uses LLM plus extra tools, memory, planning, and feebacks to generate a response. 

eg: 

prompt:"Write summary "

AI agent: 
prompt: "Research competitors, summarize findings, and draft the email "

- It breaks the prompt into steps, serches using tools, gather findings, and then write the email.

### Develop a platfrom of agentic AI for an Interiewer taking DSA Interview 

````python
import requests
ollama_url = "http://localhost:11434/api/chat"

# SYSTEM_PROMPT = """
#     You are a strict Data Structure and Algorithm interviewer. 

#     Rules: 
#     - Ask only one question at a time 
#     - start easy, then increase difficulty level. 
#     - wait for user response 
#     - Evaluate answer: 
#     - If correct -> appreciate + give better /optimal answer  and If wrong -> give hint first and then correct answer 
#     - Keep resopnse short and professional 

#     - Continue the program unit the user exit 
# """

SYSTEM_PROMPT = """
You are a strict Data Structure and Algorithm interviewer.

Rules:
- Ask only ONE question at a time
- Start from easy, then increase difficulty
- Wait for user response before next question

Evaluation Rules:
- First, analyze the user's answer carefully
- Clearly decide: CORRECT or WRONG
- If CORRECT:
    - Say "Correct"
    - Appreciate briefly
    - Provide a better or optimal solution
- If WRONG:
    - Say "Incorrect"
    - Give a small hint FIRST
    - Then provide correct answer

Style:
- Keep response short and professional
- Do NOT skip evaluation
- Do NOT ask next question until evaluation is done

Continue until user says "exit"
"""

# memory 
messages = [
     {"role":"system","content":SYSTEM_PROMPT}
 ]

def chat_with_ai(messages): 
    response = requests.post(ollama_url, json={
        "model" : "gemma3:1b", 
        "messages": messages, 
        "stream":False
    })

    return response.json()["message"]["content"]


print("Bot Interview started (type 'exit' to quit)")

messages.append({"role":"user", "content": "start interview"})
ai_reply = chat_with_ai(messages)

print("AI: ", ai_reply)

## ask question 

messages.append({"role":"assistant", "content": ai_reply})

# main loop : 

while True: 
    user_input = input("\n You: ")

    if user_input.lower() == "exit": 
        print("Interview Ended")
        break

    #store user response 

    messages.append({"role":"user","content":user_input})

    ai_reply = chat_with_ai(messages)

    print("AI", ai_reply)

    #store AI reply 
    messages.append({"role":"assistant","content":ai_reply})
    
        

````

## Working of LLM 


### Input Embedding : Convert the text in vectors. 
### Positional Encoding : It adds sequence of information in orders. 
### Self -attention: Understanding relationship between words in context. 
### Feed forward Layers: It helps to capture complex patterns. 
### Decoder: Generate response step by step 
### Parallel relationships reasoning and generates output. 

### Training Process: 

1. Pre- training : The initial phase where a model process enormous amount of text(from internet) to learn patterns.
2. Backpropagation: It is an algorithm designed to train tweak paramaters. If the model makes an incorrect predictions then
   wights are adjusted to make the correct word more likely in the future.
3. Scale of Computation: The training involves a mind-boggling numbers of operations. If a person performed one billions addition and multiplication in each seconds, then it will still take over 100 millions years to coumpute the operation for LLM.
4. Reinforcemnet learning with human feedback(RLHF): It is a secondary training type where human workers flag unhelpful prediction.So model works like a helpful assistant  rather than just text completer.  


# Tokens

    - Words : apple, entertainment 
    - Subwords : un - happy 
    - Prefix 
    - Suffix 
    - Puncuation

## Temperature : 

- Temperature in LLM is a hyperparameters that controls the randomness and creativity of the generated text. 
- Typically range between 0 to 2 
- 0 to 0.5 : lower temperature  -> deterministic, focused, factual 
- 0.7 to 1.0+ : Increases diversity and creativity , ideal for brainstroming or creating writing. 

## Key aspects: 

### How does it works:

- It shapes the probability distribution of next token. A low temperature makes the model confidently select top-ranked words. 
  A high temperature also considers words with low probability. 
- Low Temperature: Produces predictable, conservative, repetitive ouptputs. Best: Math, or facual Q&A 
- High Temperature: It increases randomness resulting more diverse, innovatinve model but probability of getting false response is high. 
    


## Context Window: 

- A context window is the span of text that a model can consider at a time while making predictions.
- It is the lookback or the amount of previous information that a model uses to make sense of current input.


- LLMs sunch as GPs and Geminis hevily rely on context windows for predicting the next token in sequence. 
- The larger the context window the more infomation model can access to understand the meaning of text.

### Let's create AI-driven student profile extractor, that takes natural language as input, Ollama as LLM, and convert all extracted data into 
### json fromat then analyze and store it  


````python
import ollama 
import json 
import os 
import re

#configuration 

MODEL_NAME = "gemma3:1b"

FILE_NAME = "students.json"

# load data from file 

def load_student():
    if (os.path.exists(FILE_NAME)): 
        try: 
            with open(FILE_NAME, "r", encoding="utf-8") as file:
                return json.load(file)
        except json.JSONDecodeError: 
            print("Warning: JSON file corrupted")
            return []
    return []

## store students record to file 

def save_student(students): 
    with open(FILE_NAME, "w", encoding= "utf-8") as file:
        json.dump(students, file, indent=2, ensure_ascii= False)

## function for retrieving record from model
def get_student_data(user_text): 
    prompt = f"""
    Act like a strict JSON generator
    Extract student record from the text 

    Return only valid JSON format 
    Do not write an explanation
    Do not write in markdown
    Do not use ```, 


    Use exactly this format: 
    {{
        "name":"", 
        "age":0, 
        "course": "", 
        "skills":[]
    }}

    Rules: 
    - age must be integer 
    - skill must be list 
    - if missing use empty string 

    Input: 
    {user_text}

    """

    response = ollama.chat(
        model=MODEL_NAME, 
        messages=[{"role":"user", "content":prompt}]
    )

    content = response["message"]["content"].strip()

    return content


## clean the model output 
def clean_json_text(text): 
    text = text.strip()

    ## case 1: If AI output is in markdown from 

    if text.startswith("```"):
        text = text.split("\n", 1)[1]

    if text.endswith("```"): 
        text = text.rsplit("\n", 1)[0]

    ## case 2: convert into exact JSON format 
    match = re.search(r"\{.*\}",text, re.DOTALL)
    if match: 
        return match.group(0)

    return text


### parse JSON output 

def parse_json_output(json_text): 
    json_text = clean_json_text(json_text)

    try: 
        data = json.loads(json_text)
        return data 
    except json.JSONDecodeError as error: 
        print("\n JSON parsing failed")
        print("Cleaned output was:")
        print(json_text)
        print(f"Error : {error}")

        return None 

## analyze 

def analyze_student(data): 
    print("\n Student analysis:")

    #extract values 

    name = data.get("name", "")
    age = data.get("age", 0)
    course = data.get("course", "")
    skills = data.get("skills", [])
    

    print(f"Name : {name}")
    print(f"Age: {age}")
    print(f"Course:{course}")
    print(f"Skill:{','.join(skills) if skills else 'No skill'}")


    # logic 

    print("Code logic:")

    #age classification 

    if(age and age<18): 
        print("Student Category: Minor")
    elif(age): 
        print("Student Category: Adult")
    else: 
        print("Student Category: Unknown")

    #skill based suggestion 

    skill_list = [s.lower() for s in skills]

    if "python" in skill_list: 
        print("Suggestion : good for AI/ Backend projects")

    if "java" in skill_list: 
        print("Suggestion: good for OOP and enterprise applications")

    if not course: 
        print("Warning: Course not found")


def main(): 
    print("Student profile extractor: ")

    students = load_student()

    if students is None:
        students = []
   

    while True: 

        #get user input 

        user_text = input("\n Enter student description or 'exit'").strip()

        if(user_text.lower() == "exit"): 
            print("\n Total saved student:", len(students))
            print("Program ended")
            break

        if not user_text: 
            print("Please enter some data")
            continue

            
        print("\n Generating structured JSON \n")

        json_text = get_student_data(user_text)

        print("Raw model outupt")
        print(json_text)

        #parse 

        data = parse_json_output(json_text)

        if data is None: 
            print("Try again with clear input")
            continue

        #analyze 

        analyze_student(data)
        students.append(data)
  
        
        save_student(students)

        print("Students record saved sucessfully !!!")


## run program 

if __name__ == "__main__":
    main()

        
        
````
